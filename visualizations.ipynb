```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_path = "C:/Users/Sparsh/Downloads/QA_3.csv"  
data = pd.read_csv(file_path, encoding='latin1')


print("Summary Statistics:")
print(data.describe())
print("\nMissing Values:")
print(data.isnull().sum())


plt.figure(figsize=(14, 6))


plt.subplot(1, 2, 1)
sns.histplot(data['BLEU Score'], bins=10, kde=True, color='skyblue')
plt.title('BLEU Score Distribution')
plt.axvline(data['BLEU Score'].mean(), color='red', linestyle='--', label=f"Mean: {data['BLEU Score'].mean():.2f}")
plt.legend()
plt.subplot(1, 2, 2)
sns.histplot(data['ROUGE Score'], bins=10, kde=True, color='salmon')
plt.title('ROUGE Score Distribution')
plt.axvline(data['ROUGE Score'].mean(), color='red', linestyle='--', label=f"Mean: {data['ROUGE Score'].mean():.2f}")
plt.legend()

plt.tight_layout()
plt.show()

complexity_means = data.groupby('Complexity')[['BLEU Score', 'ROUGE Score']].mean()
print("\nMean Scores by Complexity:")
print(complexity_means)

complexity_means.plot(kind='bar', figsize=(10, 6), color=['skyblue', 'salmon'])
plt.title('Mean BLEU and ROUGE Scores by Complexity')
plt.ylabel('Mean Score')
plt.xticks(rotation=45)
plt.show()

complexity_counts = data['Complexity'].value_counts()

plt.figure(figsize=(8, 6))
sns.barplot(x=complexity_counts.index, y=complexity_counts.values, palette='viridis')
plt.title('Question Complexity Distribution')
plt.xlabel('Complexity Level')
plt.ylabel('Count')
plt.show()
```
